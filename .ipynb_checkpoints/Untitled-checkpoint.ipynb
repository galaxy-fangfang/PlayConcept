```python
from llama_index.embeddings.ollama import OllamaEmbedding

ollama_embedding = OllamaEmbedding(
    model_name="llama2",
    base_url="http://localhost:11434",
    ollama_additional_kwargs={"mirostat": 0},
)
```

```python
#f = open("list.txt","r+")

with open('list.txt', 'r') as file:
    # Read all lines and store them in a list, stripping any trailing newline characters
    lines = [line.strip() for line in file]

# Print the list to verify the contents
print(lines)
```

```python
pass_embedding = ollama_embedding.get_text_embedding_batch(
    lines, show_progress=True
)
print(pass_embedding)
```

```python
import csv

# Define input and output file paths
input_csv_file = 'codes.csv'
#output_text_file = 'output.txt'

# Open the input CSV file and the output text file
with open(input_csv_file, 'r') as csv_file:
    # Create a CSV reader object
    csv_reader = csv.reader(csv_file)
    
    # Process each row in the CSV file
    for row in csv_reader:
        # Remove the first column
        row = row[1:]
        
        # Join the remaining columns into a single sentence
        sentence = ' '.join(row)
        
        # Write the sentence to the text file
        print(sentence)
```

```python
# import
from langchain_chroma import Chroma
from langchain_community.document_loaders import TextLoader
from langchain_community.embeddings.sentence_transformer import (
    SentenceTransformerEmbeddings,
)
from langchain_text_splitters import CharacterTextSplitter

```

```python
# load the document and split it into chunks
loader = TextLoader("list.txt")
documents = loader.load()
```

```python
# split it into chunks
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
docs = text_splitter.split_documents(documents)
```

```python
print(docs)
```

```python
# create the open-source embedding function
embedding_function = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
db = Chroma.from_documents(docs, embedding_function)
```

```python
# query it
query = "Alfred the Great"
#docs = db.similarity_search(query)
```

```python
# print results
#print(docs[0].page_content)
```
